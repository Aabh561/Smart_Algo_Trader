```python


import schedule
import time
import os

# ‚úÖ Define the full pipeline to run each morning
def run_all_modules():
    print("üîÅ Running automated Smart Algo-Trader pipeline...")

    os.system("jupyter nbconvert --to notebook --execute 01_data_fetcher_US.ipynb")
    os.system("jupyter nbconvert --to notebook --execute 02_strategy_logic.ipynb")
    os.system("jupyter nbconvert --to notebook --execute 03_backtester.ipynb")
    os.system("jupyter nbconvert --to notebook --execute 04_ml_predictor.ipynb")
    os.system("jupyter nbconvert --to notebook --execute 05_google_sheets_logger.ipynb")
    os.system("jupyter nbconvert --to notebook --execute 06_telegram_alerts.ipynb")

    print("‚úÖ All modules executed successfully.")

# ‚úÖ Schedule daily run at 09:00 AM
schedule.every().day.at("09:00").do(run_all_modules)

# ‚úÖ Keep the scheduler running
while True:
    schedule.run_pending()
    time.sleep(60)
```
