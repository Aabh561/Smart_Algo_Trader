```python
# üìä Alpha Vantage-Based Data Fetcher for US Stocks (AAPL, MSFT, GOOGL)

from alpha_vantage.timeseries import TimeSeries
import pandas as pd
import time

# ‚úÖ Step 1: Your Alpha Vantage API Key
api_key = "ZWJI493RM0QMEULJ"
ts = TimeSeries(key=api_key, output_format='pandas')

# ‚úÖ Step 2: US stock tickers (these always work)
stocks = ['AAPL', 'MSFT', 'GOOGL']

# ‚úÖ Step 3: Fetch and save each stock‚Äôs data
for stock in stocks:
    print(f"\nüì• Fetching data for {stock}...")
    try:
        data, meta_data = ts.get_daily(symbol=stock, outputsize='full')
        if data.empty:
            print(f"‚ùå No data returned for {stock}.")
            continue
        data = data.sort_index()  # Sort by date ascending
        file_name = f"{stock}_data.csv"
        data.to_csv(file_name)
        print(f"‚úÖ Data saved to: {file_name}")
    except Exception as e:
        print(f"‚ùå Error fetching {stock}: {e}")
    
    # ‚è± Avoid API rate limiting: 5 requests/min on free plan
    time.sleep(15)
```
