```python
#  Backtest Strategy on AAPL

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('AAPL_signals.csv')

df['Position'] = df['Buy'].shift(1).fillna(False).astype(int)
df['Returns'] = df['4. close'].pct_change()
df['Strategy'] = df['Returns'] * df['Position']

# Cumulative returns
cumulative = (df[['Returns', 'Strategy']] + 1).cumprod()
cumulative.plot(title='Backtest: Market vs Strategy', figsize=(10, 5))
plt.grid(True)
plt.show()
```
