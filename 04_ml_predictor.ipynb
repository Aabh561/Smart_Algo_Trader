```python
# ðŸ¤– Module 04: ML-Based Stock Movement Predictor for AAPL

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load data (from Alpha Vantage CSV)
df = pd.read_csv('AAPL_data.csv')
df = df.rename(columns=lambda x: x.strip())  # Clean column names

# âœ… Feature Engineering
df['Return'] = df['4. close'].pct_change().fillna(0)
df['Target'] = (df['Return'].shift(-1) > 0).astype(int)  # 1 if next day up, else 0
df['MA5'] = df['4. close'].rolling(5).mean().bfill()     # ðŸ”§ Fixed deprecation warning

# âœ… ML Model: Decision Tree
X = df[['4. close', 'MA5']]  # Features
y = df['Target']             # Label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
preds = clf.predict(X_test)

# âœ… Accuracy Score
acc = accuracy_score(y_test, preds)
print(f"âœ… ML Accuracy: {acc*100:.2f}%")
```
