```python
# ðŸ“¤ Module 05: Log Trade to Google Sheets using Google Service Account

import gspread
from oauth2client.service_account import ServiceAccountCredentials
from datetime import datetime

# âœ… Step 1: Define the required scopes
scope = [
    "https://spreadsheets.google.com/feeds",
    "https://www.googleapis.com/auth/spreadsheets",
    "https://www.googleapis.com/auth/drive"
]

# âœ… Step 2: Authorize using the credentials file
creds = ServiceAccountCredentials.from_json_keyfile_name("credentials.json", scope)
client = gspread.authorize(creds)

# âœ… Step 3: Open your Google Sheet by its name
# Make sure this name matches exactly (case-sensitive)
sheet = client.open("AlgoTradeLog").sheet1

# âœ… Step 4: Log a sample trade
stock = "AAPL"
signal = "BUY"
date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
price = "186.42"  # Optional: you can fetch from your CSV if needed

# âœ… Step 5: Append the trade row
sheet.append_row([stock, signal, date, price])

print("âœ… Trade successfully logged to your Google Sheet.")
```
